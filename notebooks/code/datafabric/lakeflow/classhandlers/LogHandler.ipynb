```python
import logging
```

```python
class Logger:
    def __init__(self, log_level=logging.INFO, name='databricks_logger', ):
        self.logger = logging.getLogger(name)
        self.logger.setLevel(log_level)
        self.stream_handler = logging.StreamHandler()
        formatter = logging.Formatter(fmt='%(asctime)s %(message)s', datefmt='%Y-%m-%d %H:%M:%S')
        self.stream_handler.setFormatter(formatter)
        self.stream_handler.setLevel(log_level)
        self.logger.addHandler(self.stream_handler)
        #self.logger.addHandler(logging.FileHandler('databricks_logs.log'))

    def set_level(self, level):
        self.logger.setLevel(level)
        
    def log(self, level, message):
        self.logger.log(level, message)

    def set_formatter(self, formatter):
        self.stream_handler.setFormatter(formatter)

    def log_info(self, message):
        self.logger.info(f"‚úì {message}")
    def log_warning(self, message):
        self.logger.warning(f"‚ö† {message}")
    def log_error(self, message):
        self.logger.error(f"‚ùå {message}")
    def log_debug(self, message):
        self.logger.debug(f"üêõ {message}")
    def log_critical(self, message):
        self.logger.critical(f"‚ùå‚ùå {message}")
    def log_exception(self, message):
        self.logger.exception(f"‚ùå {message}")
    def log_success(self, message):
        self.logger.info(f"‚úÖ {message}")
    def log_failure(self, message):
        self.logger.error(f"‚ùå {message}")
    def log_time(self, message):
        self.logger.warning(f"‚è∞ {message}")
        
    def log_debug_dataframe(self, df):
        if df and self.logger.isEnabledFor(logging.DEBUG):
            self.logger.info(f"df:\n{df}") 
            df.display()   
```

```python
log = Logger(logging.DEBUG)
log.stream_handler.flush()
log.log_time("Starting the notebook...")
log.log_info("This is an informational message.")
log.log_warning("This is a warning message.")
log.log_error("This is an error message.")
log.log_debug("This is a debug message.")
log.log_critical("This is a critical message.")
log.log_debug("This is a debug message.")
log.log_success("This is a success message.")
log.log_failure("This is a failure message.")
log.log_debug_dataframe(spark.range(10))
log.log_time("Ending the notebook...")
try:
  1/0
except ZeroDivisionError as e:
    log.log_exception("This is an exception message.")
log.stream_handler.close()
#log.logger.shutdown()
```
